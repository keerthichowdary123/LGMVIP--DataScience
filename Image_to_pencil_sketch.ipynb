**IMAGE TO PENCIL SKETCH USING PYTHON**

IMPORTING CV

```python
import cv2
```

from drive mounting the picture required

```python
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/pqr.jpg")
cv2_imshow(image)
```

grey colored image

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
inverted_img=255-gray_image
cv2_imshow(inverted_img)
```

```python
blurr = cv2.GaussianBlur(inverted_img,(21,21),0)
```

**PENCIL SKETCH**

```python
inverted_blurred = 255 - blurr
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

Both original and pencil sketched images

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
